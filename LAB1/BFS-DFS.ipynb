## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**..

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### Bài Tập Lập Trình: BFS trên Đồ Thị Mẫu 2
Chạy BFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (không trọng số)
graph2 = {
    'S': ['A', 'B'],
    'A': ['S', 'B', 'C'],
    'B': ['S', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C','G']
}

# Chạy BFS và in kết quả
path = bfs(graph2, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path)
````

#### Bài Tập Lập Trình: DFS trên Đồ Thị Mẫu 2
Chạy DFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (không trọng số)
graph2 = {
    'S': ['A', 'B'],
    'A': ['S', 'B', 'C'],
    'B': ['S', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C','G']
}

# Chạy DFS và in kết quả
path = dfs(graph2, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 2:", path)
````

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 7 (không trọng số)
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'B','D','E'],
    'B': ['A', 'C','E','F'],
    'C': ['S','F','G'],
    'D': ['S', 'A','E'],
    'E': ['S', 'A','B','D','F','H'],
    'F': ['D','B','C','E','G','H'],
    'G': ['C','F','H']
}

# Chạy DFS và in kết quả
path = dfs(graph7, 'S', 'H')
print("Đường đi DFS trên Đồ thị mẫu 7:", path)
````

### Thêm cạnh mới vào đồ thị mẫu 6 so sánh BFS và DFS

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    # Mỗi phần tử trong hàng đợi là (current_node, [path_list], current_total_weight)
    queue = deque([(start, [start], 0)])
    
    # Sử dụng một từ điển để lưu trữ trọng số ngắn nhất đã tìm thấy đến mỗi nút
    # Điều này giúp BFS tìm đường đi ngắn nhất trong đồ thị có trọng số dương
    shortest_weights = {start: 0}
    
    # Để lưu trữ đường đi ngắn nhất
    shortest_path = None
    min_total_weight = float('inf')

    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, current_total_weight) = queue.popleft()

        # Nếu chúng ta đã tìm thấy một đường đi ngắn hơn đến nút này, bỏ qua
        if current_total_weight > shortest_weights.get(node, float('inf')):
            continue
        
        # Nếu nút hiện tại là đích và đường đi này ngắn hơn đường đi ngắn nhất đã tìm thấy
        if node == goal:
            if current_total_weight < min_total_weight:
                min_total_weight = current_total_weight
                shortest_path = path
            # Không dừng lại ngay lập tức nếu có thể có đường đi ngắn hơn khác
            # trong các nhánh khác của BFS (do BFS duyệt theo lớp)
            continue # Tiếp tục để đảm bảo tìm đường đi ngắn nhất

        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph.get(node, []): # Sử dụng .get để xử lý các nút không có cạnh đi ra
            new_weight = current_total_weight + weight
            # Nếu đường đi mới đến nút kề ngắn hơn đường đi đã biết
            if new_weight < shortest_weights.get(neighbor, float('inf')):
                shortest_weights[neighbor] = new_weight
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], new_weight))
    
    return shortest_path, min_total_weight

def dfs_weighted(graph, start, goal):
    # Sử dụng một stack (list) cho DFS. Mỗi phần tử là (current_node, [path_list], current_total_weight)
    stack = [(start, [start], 0)]
    
    # Để lưu trữ tất cả các đường đi tìm thấy và trọng số của chúng
    all_paths_and_weights = []

    # Lặp cho đến khi stack rỗng
    while stack:
        # Lấy nút, đường đi, và tổng trọng số từ đầu stack (LIFO)
        (node, path, current_total_weight) = stack.pop()
        
        # Nếu nút hiện tại là đích
        if node == goal:
            all_paths_and_weights.append((path, current_total_weight))
            continue # Tiếp tục tìm các đường đi khác

        # Duyệt qua các nút kề của nút hiện tại
        # Sắp xếp ngược lại để các đường đi "tốt hơn" (thường là theo thứ tự bảng chữ cái hoặc ưu tiên)
        # được thăm trước nếu có nhiều lựa chọn, nhưng không ảnh hưởng đến việc tìm tất cả các đường đi.
        # Để tránh vòng lặp vô hạn trong DFS trên đồ thị có chu trình, cần kiểm tra nút đã có trong đường đi hiện tại
        for neighbor, weight in reversed(graph.get(node, [])): 
            if neighbor not in path: # Đảm bảo không tạo chu trình trong đường đi hiện tại
                new_weight = current_total_weight + weight
                stack.append((neighbor, path + [neighbor], new_weight))
                
    # Sau khi tìm thấy tất cả các đường đi, tìm đường đi có trọng số nhỏ nhất
    if not all_paths_and_weights:
        return None, 0
    
    # Sắp xếp các đường đi theo tổng trọng số để tìm đường đi ngắn nhất
    all_paths_and_weights.sort(key=lambda x: x[1])
    return all_paths_and_weights[0] # Trả về đường đi ngắn nhất được tìm thấy bằng DFS

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': [('A', 2), ('C', 5)],
    'A': [('S', 2), ('B', 3), ('D', 4)],
    'B': [('A', 3), ('E', 6)],
    'C': [('S', 5), ('D', 7), ('F', 9)],
    'D': [('A', 4), ('C', 7), ('E', 8)],
    'E': [('B', 6), ('D', 8), ('H', 10)],
    'F': [('C', 9), ('G', 12)],
    'G': [('F', 12), ('H', 15)]
}

print("--- Trước khi thêm cạnh mới ---")
# Chạy BFS và in kết quả ban đầu
path_bfs_orig, weight_bfs_orig = bfs_weighted(graph6, 'S', 'H')
print(f"Đường đi BFS trên Đồ thị mẫu 6: {path_bfs_orig}, Tổng trọng số: {weight_bfs_orig}")

# Chạy DFS và in kết quả ban đầu
path_dfs_orig, weight_dfs_orig = dfs_weighted(graph6, 'S', 'H')
print(f"Đường đi DFS trên Đồ thị mẫu 6: {path_dfs_orig}, Tổng trọng số: {weight_dfs_orig}")

# Thêm cạnh mới vào Đồ thị mẫu 6
graph6_modified = {
    'S': [('A', 2), ('C', 5)],
    'A': [('S', 2), ('B', 3), ('D', 4)],
    'B': [('A', 3), ('E', 6), ('H', 20)], 
    'C': [('S', 5), ('D', 7), ('F', 9)],
    'D': [('A', 4), ('C', 7), ('E', 8)],
    'E': [('B', 6), ('D', 8), ('H', 10)],
    'F': [('C', 9), ('G', 12)],
    'G': [('F', 12), ('H', 15)]
}

print("\n--- Sau khi thêm cạnh B-H với trọng số 20 ---")
# Chạy lại BFS với đồ thị đã sửa đổi
path_bfs_new, weight_bfs_new = bfs_weighted(graph6_modified, 'S', 'H')
print(f"Đường đi BFS trên Đồ thị mẫu 6 (mới): {path_bfs_new}, Tổng trọng số: {weight_bfs_new}")

# Chạy lại DFS với đồ thị đã sửa đổi
path_dfs_new, weight_dfs_new = dfs_weighted(graph6_modified, 'S', 'H')
print(f"Đường đi DFS trên Đồ thị mẫu 6 (mới): {path_dfs_new}, Tổng trọng số: {weight_dfs_new}")
````
